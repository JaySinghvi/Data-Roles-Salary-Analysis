```python
import pandas as pd
import numpy as np
```

```python
sal = pd.read_csv("data_science_salaries.csv")
sal_df = pd.DataFrame(sal)
sal_df.info()
```

```python
sal_df.shape
```

```python
sal_df.describe()
```

```python
sal_df.head()
```

```python
sal_df = sal_df.drop(columns = "salary")
sal_df = sal_df.drop(columns = "salary_currency")
sal_df = sal_df.drop(columns = "employee_residence")
sal_df
```

```python
sal_df.isna().sum()
```

```python
sal_df = sal_df[~sal_df["employment_type"].isin(["Contract", "Freelance", "Part-time"])]
sal_df = sal_df[sal_df["company_location"] == "United States"]
sal_df
```

```python
sal_df = sal_df.drop(columns = "company_location")
sal_df = sal_df.drop(columns = "employment_type")
sal_df
```

```python
sal_df = sal_df.rename(columns = {"work_models" : "Work_Mode"})
sal_df = sal_df.rename(columns = {"job_title" : "Job_Title", "experience_level" : "Experience", "work_year" : "Year", "salary_in_usd" : "Salary", "company_size" : "Company_Size"})
```

```python
sal_df
```

```python
sal_df["Job_Title"] = sal_df["Job_Title"].replace("Machine Learning Engineer", "ML Engineer")
```

```python
sal_df["Job_Title"].value_counts()
```

```python
sal_df = sal_df[sal_df["Job_Title"].isin(["Data Scientist", "Data Engineer", "Data Analyst", "ML Engineer", "AI Scientist"])]
```

```python
sal_df
```

```python
sal_df.reset_index(drop = True, inplace = True)
```

```python
sal_df
```

```python
sal_df.to_csv("salaries.csv", index=False)
```
